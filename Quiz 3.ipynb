```python
import pandas as pd
bel_dict = pd.read_csv("belize_airbnb_listings.csv")
bel_df = pd.DataFrame(bel_dict)
bel_df
grouped_df = bel_df.groupby("neighbourhood")
grouped_bel_df = bel_df.groupby("neighbourhood")


my_price = grouped_bel_df["price"].mean()

my_df_2 = pd.DataFrame(my_price).reset_index()


my_aval = grouped_bel_df["availability_365"].max()

my_df_3 = pd.DataFrame(my_aval).reset_index()


my_rev = grouped_bel_df["number_of_reviews"].sum()

my_df_4 = pd.DataFrame(my_rev).reset_index()



outer_merge_df = pd.merge(my_df_2, my_df_3, on="neighbourhood")
outer_merge_df
final_df = pd.merge(outer_merge_df, my_df_4, on="neighbourhood")
final_df
```
